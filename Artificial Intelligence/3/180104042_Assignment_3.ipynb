**1. Implement Genetic Algorithm for 8 Queens problem using Python. For the mutation
step, use inversion mutation.**

```python
%reset
def horizontal(pos):
    for i in range(len(pos)):
        countHorizontal(i, pos)

def diagonalUp(pos):
    for i in range(len(pos)):
        countDiagonalUp(i, pos)

def diagonalDown(pos):
    for i in range(len(pos)):
        countDiagonalDown(i, pos)

def countHorizontal(point, pos):
    global count
    for i in range(point + 1, len(pos)):
        if pos[i] == pos[point]:
            count += 1

def countDiagonalUp(point, pos):
    global count
    for i in range(point + 1, len(pos)):
        if pos[i] == pos[point] + i - point:
            count += 1

def countDiagonalDown(point, pos):
    global count
    for i in range(point + 1, len(pos)):
        if pos[i] == pos[point] - i + point:
            count += 1

count=0
```

```python
import random

dic=[]
for i in range(0,15):
  x=[]
  for i in range(0,8):
    x.append(random.randint(0,7))
  horizontal(x)
  diagonalUp(x)
  diagonalDown(x)
  dic.append((count,x))
  count=0

while(1):
  #print(dic)
  dic=sorted(dic)
  if dic[0][0]==0:
    break
  #print(dic)

  parent1=dic[0]
  parent2=dic[1]
  #print(parent1[1],parent2[1])
  cross1=[]
  cross2=[]

  for i in range(0,8):
    if i<4:
      cross1.append(parent1[1][i])
      cross2.append(parent2[1][i])

  for i in range(0,4):
      cross1.append(parent2[1][i+4])
      cross2.append(parent1[1][i+4])

  #crossover done
  #print(cross1,cross2)

  #mutation
  pos1=random.randint(0,7)
  pos2=random.randint(0,7)
  #value1=random.randint(0,7)
  #value2=random.randint(0,7)
  #cross1[pos1]=value1
  #cross2[pos2]=value2
  print(pos1,pos2)
  ##reverse
  if pos1 > pos2:
      pos1, pos2 = pos2, pos1
  # Reverse array[i:j+1]
  section = cross1[pos1:pos2+1]
  section.reverse()
  cross1[pos1:pos2+1] = section
  # Reverse array[i:j+1]
  section = cross2[pos1:pos2+1]
  section.reverse()
  cross2[pos1:pos2+1] = section

  print(cross1,cross2)

  #again genetic
  horizontal(cross1)
  diagonalUp(cross1)
  diagonalDown(cross1)
  dic[0]=(count,cross1)
  count=0
  horizontal(cross2)
  diagonalUp(cross2)
  diagonalDown(cross2)
  dic[1]=(count,cross2)
  count=0

print(dic)
```

**2. Implement Greedy Best First search algorithm using Python.**

```python
from queue import PriorityQueue

closed = []
open = []
value = {}
pair = []
q = PriorityQueue()

def printf():
  print(f'Closed list: {closed}')
  print(f'Open list: {open}')
  print(f'The length of the path is: {cost}')

def gbfs(begin, end):
    if begin==end:
      printf()

    else:
      for i in value:
        if i[0]==begin and i[1] not in closed and i[1]!=end:
          q.put((value[i], i[1]))
        elif i[0]==begin and i[1] not in closed and i[1]==end:
          q.put((value[i], i[1]))
          break
        #print(list(q.queue))
      if not q.empty():
        next_item = q.get()
        closed.append(begin)
        open.append(next_item[1])
        global cost
        cost+=value[(begin,next_item[1])]
        gbfs(next_item[1], end)


def makePair(x, y):
  temp=(x,y)
  pair.append(temp);

def makeValue(x, y, cost):
  value[x,y]=cost
  makePair(x,y)

makeValue('S','A',7)
makeValue('S','B',2)
makeValue('S','C',3)
makeValue('A','B',3)
makeValue('A','D',4)
makeValue('B','D',4)
makeValue('B','H',1)
makeValue('D','F',5)
makeValue('H','F',3)
makeValue('H','G',2)
makeValue('G','E',2)
makeValue('C','L',2)
makeValue('L','I',4)
makeValue('L','J',4)
makeValue('I','K',4)
makeValue('J','K',4)
makeValue('K','E',5)

print(value)

begin = str(input('Enter Begining node: '))
end = str(input('Enter Ending node: '))
gbfs(begin, end)
```
